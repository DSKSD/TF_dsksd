```python
import tensorflow as tf
```

### normal loading 

```python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
z = tf.add(x,y)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    writer = tf.summary.FileWriter('./graph/l2', sess.graph)
    for _ in range(10):
        sess.run(z)
        print(tf.get_default_graph().as_graph_def())
        writer.close()
```

### lazy loading 

```python
x = tf.Variable(10, name='x')
y = tf.Variable(20, name='y')
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    writer = tf.summary.FileWriter('.graph/l2', sess.graph)
    
    for _ in range(10):
        sess.run(tf.add(x,y)) # 미리 Computational graph를
        # 만들어두지 않고 세션에서 불러오게 되면 : lazy loading
        
    print(tf.get_default_graph().as_graph_def())
    writer.close()
```

lazy loading을 하게 되면 보시다시피 그래프가 보기 안좋게 부푼다

### tf.get_default_graph() 

현재 스레드의 디폴트 그래프를 리턴한다. <br>
Graph.as_default()가 명시적으로 입력되면, 리턴된 그래프는 innermost graph가 될 것이다?
